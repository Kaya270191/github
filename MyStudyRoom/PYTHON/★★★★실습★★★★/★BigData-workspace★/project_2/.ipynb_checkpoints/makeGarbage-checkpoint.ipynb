#### 필요한 구당 쓰레기 발생 합계데이터를 추출하기 위한 반복문
### 본인 구당 42개의 row를 가지고있음
### 처리된 row를 저장하기위한 dataframe생성

```python
NewmakeGarbage=pd.DataFrame(columns=['시도','시군구','발생량'])
NewmakeGarbage
```

```python
num=0
for i in range(1,len(makeGarbage),42):
    district=makeGarbage.loc[i]
    NewmakeGarbage.loc[num]=[district['시도'],district['시군구'],district['2019년\n발생량']]
    print(district['시도'])
    print(district['시군구'])
    print(district['2019년\n발생량'])
    num=num+1
```
