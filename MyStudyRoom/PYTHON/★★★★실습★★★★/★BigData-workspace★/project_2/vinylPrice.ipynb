```python
def renamesido(x):
    if(x=='경기도'):
        return '경기'
    if(x=='서울특별시'):
        return '서울'
    if(x=='전라북도'):
        return '전북'
    if(x=='강원도'):
        return '강원'
    if(x=='인천광역시'):
        return '인천'
    if(x=='경상북도'):
        return '경북'
    if(x=='경상남도'):
        return '경남'
    if(x=='부산광역시'):
        return '부산'
    if(x=='충청남도'):
        return '충남'
    if(x=='전라남도'):
        return '전남'
    if(x=='충청북도'):
        return '충북'
    if(x=='울산광역시'):
        return '울산'
    if(x=='대전광역시'):
        return '대전'
    if(x=='대구광역시'):
        return '대구'
    if(x=='광주광역시'):
        return '광주'
    if(x=='제주특별자치도'):
        return '제주'
    if(x=='세종특별자치시'):
        return '세종'
```

```python
vinylPrice['시도명']=vinylPrice['시도명'].apply(renamesido)
vinylPrice
```

```python
#쓰레기 봉투가 아닌 마대와 김장용봉투 인 row 제거
vinylPrice2=vinylPrice[vinylPrice['종량제봉투종류']!='특수규격마대']
vinylPrice3=vinylPrice2[vinylPrice2['종량제봉투종류']!='김장용봉투']
vinylPrice4=vinylPrice3[vinylPrice3['종량제봉투사용대상']!='가정용(공사장생활폐기물)']
```

```python
vColumns=vinylPrice4.columns
```

```python
#mean을 하기 위해 0인 데이터를 None으로 처리
for column in vColumns:
    vinylPrice3[column]=vinylPrice3[column].apply(lambda x : None if(x==0) else x)
```

```python
vinylPrice5=vinylPrice4.groupby(['시도','시군구']).mean()
```
