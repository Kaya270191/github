```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from bs4 import BeautifulSoup
import re #정규식 표현을 위한 모듈 
import requests
```

```python
titles=[] #타이틀만 담은 리스트
hrefs=[] #링크만 담은 리스트 
contentList=[] #컨텐츠리스트만 담은 리스트 

#데이터 프레임
columns = ["title", "hrefs" , "contentList"]
df = pd.DataFrame(columns=columns)


base_url='https://www.kita.net/cmmrcInfo/rsrchReprt/ovseaMrktReprt/ovseaMrktReprtDetail.do?pageIndex=1&no='
base_url2='&type='




for num in range(2):
    response = requests.get(f"https://www.kita.net/cmmrcInfo/rsrchReprt/ovseaMrktReprt/ovseaMrktReprtList.do?pageIndex={num}")
    response
    html = response.text
    soup = BeautifulSoup(html,'html.parser')
    
    #제목 추출
    aTag = soup.select('#contents > div.boardArea > div.boardList.type2 > ul > li > a')
    for a in aTag:
        title = a.text.replace('\t','').replace('\n','') #제목 추출
        titles.append(title) # 제목 리스트에 추가
        
        href = a.attrs['href'].split(',')[1] #링크 
        href_type=aTag[0].attrs['href'].split(',')[3]
        href_type=re.sub(r'[^0-9]', '',href_type) #숫자만 추출
        href = base_url+href+base_url2+href_type
        href = href.replace('\n','')
        response= requests.get(href)
        html = response.text
        soup = BeautifulSoup(html,'html.parser')
        if soup.select('#contents > div.boardArea > div.boardView > div.txtArea > p'):    
            con=[]
            content=soup.select('#contents > div.boardArea > div.boardView > div.txtArea > p')
            for i in content:
                con.append(i.text)  
            contentList.append(''.join(con))
        else:
            contentList.append(" ")
            
 #데이터프레임           
    row = [titles, hrefs, contentList]
    series = pd.Series(row, index=df.columns)
    df = df.append(series, ignore_index=True)
    
df
```

```python
contentList=[]
for num in hrefs:
    response= requests.get(num)
    html = response.text
    soup = BeautifulSoup(html,'html.parser')
    content=soup.select('#contents > div.boardArea > div.boardView > div.txtArea > p')

    for i in content:    
        content=i.text
        contentList.append(content)
        
    
```

```python
#데이터 프레임 만들기
columns = ["title", "hrefs" , "contentList"]
df = pd.DataFrame(columns=columns)


for i in range(20): 
    row = [titles[i], hrefs[i], "".join(contentList[i])]
    series = pd.Series(row, index=df.columns)
    df = df.append(series, ignore_index=True)

df

```
